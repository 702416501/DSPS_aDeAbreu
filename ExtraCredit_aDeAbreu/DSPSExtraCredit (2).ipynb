#Task 1: Chi Squared Tests on the observed and predicted data 

```
def pvalue(test):
  '''
  parameter:
  test->results of the chisquare test

  return:
  pvalue of the the test

  '''
  return np.exp(-1 * (test.statistic/2))
```

```
midwaytest1 = sp.stats.chisquare(midwayrelevant[owl], midwayrelevant[pwl], ddof = midwayrelevant.shape[0] -1)
nawiltest1 = sp.stats.chisquare(nawilrelevant[owl], nawilrelevant[pwl], ddof = nawilrelevant.shape[0] - 1)
honotest1 = sp.stats.chisquare(honorelevant[owl], honorelevant[pwl], ddof = honorelevant.shape[0] -1)
kahultest1 = sp.stats.chisquare(kahulrelevant[owl], kahulrelevant[pwl], ddof = kahulrelevant.shape[0] - 1)
kawaitest1 = sp.stats.chisquare(kawairelevant[owl], kawairelevant[pwl], ddof = kawairelevant.shape[0] - 1)
hilotest1 = sp.stats.chisquare(hilorelevant[owl], hilorelevant[pwl], ddof = hilorelevant.shape[0] - 1)

task1tests = [midwaytest1, nawiltest1,honotest1, kahultest1, kawaitest1, hilotest1]
gauges = ["Midway Islands", "Nawiliwili", "Honolulu",  "Kahului", "Kawaihae", "Hilo"]
```

```
for i in range(6):
  print("The chi square test statistic for {} is {:.2f} and the p-value is {:.2}".format(gauges[i], task1tests[i].statistic, pvalue(task1tests[i])))
  if pvalue(task1tests[i]) < .05:
    print("Bescause the the p-value < 0.05, we HAVE sufficient evidence to reject the null\nhypothesis that the predicted data is a good fit for the observed data")
  else:
    print("Becasue the p-value >= 0.05, we DO NOT HAVE sufficient evidence to reject the null\nhypothesis that the predicted data is a good fit for the observed data")

  print("\n**************************************************************************\n")
```
